```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2-3*np.random.normal(0,1,20)
Y = 10-2*X**2 + 0.5*X**3 +np.random.normal(-3,3,20)
```

```python
def regression(X,k):
    if k <1:
        X_ls=np.array(20*[1]).reshape(20,1)
    else:
        X_ls=np.array(20*[1]).reshape(20,1)
        for i in range(k):
            X_ls=np.concatenate((X_ls,(X**(i+1)).reshape(20,1)),axis=1)
    
    beta = np.linalg.inv(X_ls.T@X_ls)@(X_ls.T)@Y
    Y_hat = X_ls @ beta
    return Y_hat
```

```python
regression(X,6)
```

```python
errors =[]
plt.figure(figsize=(18, 9))
for k in range(1,7):
    y_hat = regression(X,k)
    errors.append(np.linalg.norm(Y-y_hat)**2)
    plt.subplot(2,3,k)
    
    plt.title(f' k = {k}')
    plt.plot(X,Y,'go')
    plt.plot(X,y_hat,'ro')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(['Y','Y_hat'])
```

```python
k=[1,2,3,4,5,6]
plt.plot(k,errors,'r-')
plt.title('Error function')
plt.xlabel('k')
plt.ylabel('$\sum_{i=1}^{n} ||y - y_ls||^2$')
```
