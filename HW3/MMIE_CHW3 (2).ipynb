ÿ®ÿß ÿ™Ÿàÿ¨Ÿá ÿ®Ÿá ŸÜ⁄©ÿßÿ™ ÿ∞⁄©ÿ± ÿ¥ÿØŸá ÿØÿ± ŸÇÿ≥ŸÖÿ™ ŸÇÿ®ŸÑ ÿØÿ± ŸÇÿ≥ŸÖÿ™ ÿ≤€åÿ± ÿ™ÿßÿ®ÿπ€å ÿ±ÿß ÿ®ŸÜŸà€åÿ≥€åÿØ ⁄©Ÿá ÿ®ÿß ⁄Øÿ±ŸÅÿ™ŸÜ ŸÖÿßÿ™ÿ±€åÿ≥ ŸÖÿ±ÿ®ÿπ€å A €å⁄© ŸÖÿßÿ™ÿ±€åÿ≥ ŸÇÿ∑ÿ±€å ŸÖÿ™ÿ¥ÿßÿ®Ÿá ÿ®ÿß ÿ¢ŸÜ €åÿπŸÜ€å D Ÿà €å⁄© ŸÖÿßÿ™ÿ±€åÿ≥ V ÿßÿ≤ ÿ®ÿ±ÿØÿßÿ±Ÿáÿß€å Ÿà€å⁄òŸá ŸÖÿ™ŸÜÿßÿ∏ÿ± ÿ®ÿß D ÿ±ÿß ÿ®ÿ±⁄Øÿ±ÿØÿßŸÜÿØ.

```python
# Complete this part!
def eigen(A):
  # input: square matrix A
  # output: diagonal matrix D & eigenvector matrix V
  # NOTE: the equation "A = V @ D @ np.linalg.inv(V)" must hold 
    n = A.shape[0]
    eigenvalues = []
    x =np.array([1,2,3])
    
    x50 = np.linalg.matrix_power(A , 50) @ x / np.linalg.norm(np.linalg.matrix_power(A , 50) @ x)
    Lambda = ( A @ x50 / x50 )[0]
    eigenvalues.append(Lambda)
    vi = x50
    V  = vi.reshape(n,1)
    
    for i in range(n-1):
        A = A -  Lambda * x50.reshape(3,1) @ x50.reshape(1,3)
        x50 = np.linalg.matrix_power(A , 50) @ x / np.linalg.norm(np.linalg.matrix_power(A , 50) @ x)
        Lambda = ( A @ x50 / x50 )[0]
        eigenvalues.append(Lambda)
        vi = x50
        V  = np.concatenate((V ,vi.reshape(n,1) ),axis=1)            
    D = np.diag(eigenvalues)
    return D, V
```
